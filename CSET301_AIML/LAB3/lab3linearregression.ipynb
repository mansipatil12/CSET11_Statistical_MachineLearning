```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
experience = np.array([16, 12, 18, 4, 3, 10, 5, 12])
performance_rating = np.array([87, 88, 89, 68, 78, 80, 75, 83])
```

(i)

```python
n = len(experience)
sum_x = np.sum(experience)
sum_y = np.sum(performance_rating)
sum_xy = np.sum(experience * performance_rating)
sum_x_squared = np.sum(experience**2)
slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
intercept = (sum_y - slope * sum_x) / n
print("Slope:", slope)
print("Intercept:", intercept)
```

(ii)

```python
plt.scatter(experience, performance_rating, label='Data Points')

x_values = np.linspace(min(experience), max(experience), 100)
y_values = slope * x_values + intercept
plt.plot(x_values, y_values, color='red', label='Best-Fitted Line')

plt.xlabel('Experience (years)')
plt.ylabel('Performance Rating')
plt.title('Best-Fitted Line using Least Squares Method')
plt.legend()
plt.grid(True)
plt.show()
```

(iii)

```python
estimated_performance_20_years = slope * 20 + intercept
print("Estimated Performance Rating for 20 years of experience:", estimated_performance_20_years)
```
