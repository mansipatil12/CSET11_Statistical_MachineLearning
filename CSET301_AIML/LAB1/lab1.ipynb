1. Tabular, Spreadsheet and Interchange Data Formats

(a) Table (.arff)

```python
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
```

```python
def read_arff(file_path):
    data, meta = arff.loadarff(file_path)
    df = pd.DataFrame(data) 
    return df
```

```python
def print_properties(df):
    print("Size of the data:", df.size)
    print("Shape of the data:", df.shape)
    print("Dimensions of the data:", df.ndim)
    print("Data types of columns:\n", df.dtypes)
```

```python
def visualize_data(df):
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    for col in numeric_columns:
        plt.figure(figsize=(8, 6))
        plt.hist(df[col], bins=20, color='green', edgecolor='black')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
```

```python


def visualize_data_line(df):
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    for col in numeric_columns:
        plt.figure(figsize=(8, 6))
        plt.plot(df[col], color='blue')
        plt.title(f'Line Plot of {col}')
        plt.xlabel('Index')
        plt.ylabel(col)
        plt.show()

```

```python
def visualize_data_scatter(df):
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Ensure there are at least 2 numeric columns to create scatter plots
    if len(numeric_columns) >= 2:
        for i in range(len(numeric_columns) - 1):
            plt.figure(figsize=(8, 6))
            plt.scatter(df[numeric_columns[i]], df[numeric_columns[i + 1]], color='red')
            plt.title(f'Scatter Plot of {numeric_columns[i]} vs {numeric_columns[i + 1]}')
            plt.xlabel(numeric_columns[i])
            plt.ylabel(numeric_columns[i + 1])
            plt.show()
    else:
        print("Not enough numeric columns for scatter plots.")
```

```python
file_path = 'zoo.arff'
data_frame = read_arff(file_path)
print_properties(data_frame)
```

```python
visualize_data(data_frame)
visualize_data_line(data_frame)
```

```python
from PIL import Image

def convert_to_black_and_white(input_image_path, output_image_path):
    original_image = Image.open(input_image_path)
    bw_image = original_image.convert("L")
    bw_image.save(output_image_path)
input_path = "sunset.jpg"
output_path = "sunset_bw.jpg"

convert_to_black_and_white(input_path, output_path)

image_path = 'sunset.jpg'
img = mpimg.imread(image_path)

plt.imshow(img)
plt.axis('off')  
plt.show()
```

(b) 3D images (DIACOM, MHA)

```python
import pydicom
import matplotlib.pyplot as plt
```

```python
def read_and_visualize_dicom(file_path):
    dicom_data = pydicom.dcmread(file_path)
    pixel_array = dicom_data.pixel_array
    plt.imshow(pixel_array, cmap='gray')
    plt.title('DICOM Image')
    plt.show()
```

```python
dicom_file_path = "56364398.dcm"
read_and_visualize_dicom(dicom_file_path)
```

4. Video Data Formats

(a) MP4

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def read_and_visualize_mp4(file_path):
    cap=cv2.VideoCapture(file_path)
    if not cap.isOpened():
        print("Error opening video file")
        return
    while True:
        ret, frame= cap.read()
        if not ret:
            break
        frame_rgb=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame_rgb)
        plt.title('Video Frame')
        plt.pause(0.03)  # Pause for a short time to display the frame
        plt.clf()  # Clear the figure for the next frame
        cap.release()

video_file_path = "sample.mp4"
read_and_visualize_mp4(video_file_path)
```

5. Audio Data Formats

(a) MP3

```python
from pydub import AudioSegment
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_and_visualize_mp3(file_path):
    audio=AudioSegment.from_file(file_path, format="mp3")
    samples=np.array(audio.get_array_of_samples())
    time=np.arrange(0, len(samples))/audio.frame_rate

    plt.plot(time, samples) 
    plt.title('Audio Waveform')
    plt.xlabel('Time in seconds')
    plt.ylabel('Amplitude')
    plt.show()


    
```

```python
mp3_file_path="Night_Changes.mp3"
read_and_visualize_mp3(mp3_file_path)
```

6. Text Data Formats

(a) TXT/PDF/DOC

```python
import matplotlib.pyplot as plt

def read_and_print_txt(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print("File content:")
            print(content)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"Error: An unexpected error occurred - {str(e)}")

# Example usage:
txt_file_path = "sample.txt"
read_and_print_txt(txt_file_path)
```
