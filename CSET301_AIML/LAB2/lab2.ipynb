```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset.csv')
```

Q.1 Read Total profit of all months and show it using a line plot.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.show()
```

Q.2 Get total profit of all months and show line plot 

```python
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='blue', marker='o', markerfacecolor='red', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.legend(loc='lower right')
plt.show()
```

Q.3 Multiple line plot

```python
fig, ax = plt.subplots(figsize=(10, 6))
for product in df.columns[1:7]:
    ax.plot(df['month_number'], df[product], label=product)

ax.set(xlabel='Month Number', ylabel='Units Sold', title='Product Sales Data - Multiline Plot')
ax.legend(loc='lower right')    
plt.show()
```

Q.4 Scatter plot for toothpaste sale


```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['month_number'], df['toothpaste'])
ax.set(xlabel='Month Number', ylabel='Units Sold', title='Toothpaste Sales Data')
ax.grid(True)
plt.show()
```

Q.5 Bar Graph for Face cream and Facewash

```python
plt.figure(figsize=(10, 6))
bar_width = 0.35
bar_positions_facecream = df['month_number']
bar_positions_facewash = df['month_number'] + bar_width
plt.bar(bar_positions_facecream, df['facecream'], label='Face Cream', width=bar_width, color='blue')
plt.bar(bar_positions_facewash, df['facewash'], label='Face Wash', width=bar_width, color='orange')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Facewash Sales Data')
plt.legend()
plt.show()
```

Q.6 Bar chart for Bathing Soap and save to hard disk

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df['month_number'], df['bathingsoap'])
ax.set(xlabel='Month Number', ylabel='Units Sold', title='Bathing Soap Sales Data - Bar Chart')
plt.savefig('bathing_soap_sales.png')
plt.show()
```

Q.7 Histogram for total profits


```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['total_profit'], bins=10, edgecolor='blue')
ax.set(xlabel='Total Profit', ylabel='Frequency', title='Histogram of Total Profit')
plt.show()
```

Q.8 Pie chart for Total Sale Data

```python
total_sale = df.iloc[:, 1:7].sum()
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(total_sale, labels=total_sale.index, autopct='%1.1f%%', startangle=140)
ax.set(title='Total Sale Data for Last Year')
plt.show()
```

Q.9 Subplot for Bathing Soap and Facewash

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot(df['month_number'], df['bathingsoap'], marker='o')
ax1.set(xlabel='Month Number', ylabel='Units Sold', title='Bathing Soap Sales Data')
ax2.plot(df['month_number'], df['facewash'], marker='o')
ax2.set(xlabel='Month Number', ylabel='Units Sold', title='Facewash Sales Data')
plt.tight_layout()
plt.show()
```

Q.10 Stack plot for all product sales data

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'], labels=df.columns[1:7])
ax.set(xlabel='Month Number', ylabel='Units Sold', title='Stack Plot - Product Sales Data')
ax.legend(loc='upper left')
plt.show()
```
