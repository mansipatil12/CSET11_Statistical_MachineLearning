```python
import pandas as pd
df1=pd.read_csv('winequality-red.csv',sep=';')
df1
```

```python
df1.columns
independent=df1[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']]
independent

```

```python
dependent=df1['quality']
dependent

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def categorize_quality(score):
    if score >= 7:
        return 'Best'
    elif score >= 5:
        return 'Average'
    else:
        return 'Poor'

dependent = dependent.apply(categorize_quality)
dependent
```

```python
#split training and testing data in 75-25 ratio
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
y=df1[['quality']].copy()
features=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
x=df1[features].copy()

X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent, train_size=0.75,test_size=0.25, random_state=40)
X_train
```

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, Y_train)
y_pred = rf_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
c=confusion_matrix(Y_test,y_pred)
print(c)
```

```python
recall = recall_score(Y_test, y_pred, average='weighted')
recall
```

```python
f1 = f1_score(Y_test, y_pred, average='weighted')
f1
```

```python
#predict the target values in testing set
from sklearn.ensemble import RandomForestClassifier
rf1_model = RandomForestClassifier(n_estimators=200,criterion='gini',max_depth=12,min_samples_split=12,min_samples_leaf=12,max_features='log2',random_state=42)
rf1_model.fit(X_train, Y_train)
y_pred1 = rf1_model.predict(X_test)
y_pred1
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
c1=confusion_matrix(Y_test,y_pred1)
print(c1)
```

```python
accuracy1 = accuracy_score(Y_test, y_pred1)*100
accuracy1
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
#As the deafaul value is gini ,if we dont't specify it will select gini criteria by default
clf =DecisionTreeClassifier(criterion="gini",max_leaf_nodes=5,random_state=0)
clf.fit(X_train,Y_train)
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
y_predicted=clf.predict(X_test)
print(y_predicted)

```
