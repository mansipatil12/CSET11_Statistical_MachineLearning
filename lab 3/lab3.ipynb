```python
# Display all the features in titanic data
print(titanic_df.columns)
```

```python
# Find out the number of examples of each class passenger
passenger_counts = titanic_df['Pclass'].value_counts()
print(passenger_counts)
```

```python
import matplotlib.pyplot as plt
```

```python
# a. Bar Chart
titanic_df['Sex'].value_counts().plot(kind='bar')
plt.title('Passenger Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# b. Multi Line Chart
ages = titanic_df['Age'].dropna()
plt.plot(ages, marker='o')
plt.title('Age Distribution')
plt.xlabel('Passenger')
plt.ylabel('Age')
plt.show()
```

```python
# c. Line Chart
titanic_df.groupby('Pclass')['Fare'].mean().plot(marker='o')
plt.title('Average Fare by Class')
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.show()
```

```python
# d. Histogram with Color Fill
titanic_df['Age'].plot(kind='hist', edgecolor="black", bins=20, color='purple', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# e. Box Plot
titanic_df.boxplot(column='Fare', by='Pclass')
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
# f. Scatter Plot (2D and 3D)
plt.scatter(titanic_df['Age'], titanic_df['Fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(titanic_df['Pclass'], titanic_df['Age'], titanic_df['Fare'])
ax.set_xlabel('Class')
ax.set_ylabel('Age')
ax.set_zlabel('Fare')
plt.title('3D Scatter Plot')
plt.show()
```

```python
# g. Scatter Plot with Different Colors for Each Class
colors = ['red', 'green', 'blue']
for i, pclass in enumerate(titanic_df['Pclass'].unique()):
    subset = titanic_df[titanic_df['Pclass'] == pclass]
    plt.scatter(subset['Age'], subset['Fare'], c=colors[i], label=f'Class {pclass}')
plt.title('Age vs Fare with Class Distinction')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
# h. Box Plot for Individual Features
titanic_df.boxplot(column=['Age', 'Fare'], by='Pclass', layout=(1, 2))
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# i. Violin Plot
import seaborn as sns
sns.violinplot(x='Pclass', y='Age', data=titanic_df)
plt.title('Age Distribution by Class (Violin Plot)')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
# j. KDE Plot
sns.kdeplot(titanic_df['Age'], shade=True)
plt.title('Kernel Density Estimation of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
# k. Calculate the Correlation Matrix
corr = titanic_df.corr()
```

```python
# l. Heatmap Plot for Correlation Matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
