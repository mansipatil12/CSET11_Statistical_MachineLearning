```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
XY = np.genfromtxt("./winequality-white.csv", delimiter=";", skip_header=1)
```

```python
#i
print(XY.shape)
```

```python
#ii a
Y = XY[:, -1]
print("Shape of Y:", Y.shape)
```

```python
#ii b
X = XY[:, :-1]
print("Shape of X:", X.shape)
```

```python
#ii c
np.random.shuffle(X)  # Shuffling the rows of X
X_transposed = X.T #transpose shuffled array
print("Shape of X transposed:", X_transposed.shape)
```

```python
#ii d
max_values_across_columns = np.max(X, axis=0)
min_values_across_columns = np.min(X, axis=0)
max_values_across_rows = np.max(X, axis=1)
min_values_across_rows = np.min(X, axis=1)
print("Max values across columns:", max_values_across_columns)
print("\nMin values across columns:", min_values_across_columns)
print("\nMax values across rows:", max_values_across_rows)
print("\nMin values across rows:", min_values_across_rows)
```

```python
#ii e
count_equal_to_5 = np.count_nonzero(Y == 5)
print("Number of values equal to 5 in Y:", count_equal_to_5)
```

```python
#Q4 i
mean_columns_X = np.mean(X, axis=0)
print("Mean for all columns in X:", mean_columns_X)
```

```python
#Q4 ii
mode_last_column_Y = np.argmax(np.bincount(Y.astype(int)))
print("Mode of the last column (Y):", mode_last_column_Y)
```

```python
#Q.4 iii
stddev_columns_X = np.std(X, axis=0)
print("Standard deviation for all columns in X:", stddev_columns_X)
```

```python
#Q.5 i
unique_values_Y, counts = np.unique(Y, return_counts=True)

plt.figure(figsize=(8, 6))
plt.bar(unique_values_Y, counts, color='blue')
plt.title('Bar Graph of Unique Values in Y')
plt.xlabel('Quality of Wine (Y)')
plt.ylabel('Frequency')
plt.xticks(unique_values_Y)
plt.show()
```

```python
#Q.5 ii
num_columns_X = X.shape[1]
num_rows = (num_columns_X + 1) // 2  # Calculate the number of rows needed

plt.figure(figsize=(12, 8))
for column_index in range(num_columns_X):
    plt.subplot(num_rows, 2, column_index + 1)  # Use num_rows and 2 for the subplot layout
    plt.hist(X[:, column_index], bins=20, color='green', alpha=0.7)
    plt.title(f'Histogram for Column {column_index}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
